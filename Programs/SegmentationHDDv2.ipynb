```python
#https://pytorch.org/tutorials/beginner/blitz/cifar10_tutorial.html#training-on-gpu
import torch, torchvision, os, cv2, random
from torch.utils.data import random_split, Dataset
from torch.nn import Tanh, Linear, ReLU, Sequential, Conv2d, MaxPool2d, Sigmoid, BatchNorm2d, Flatten, ConvTranspose2d
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
from torchvision import models

class augementData(Dataset):

    def __init__(self, data, transform):
        self.data = data
        self.transform = transform

    def __len__(self):
        return len(self.data.shape[0])

    def __getitem__(self, idx):
        item = self.data[idx]
        item = self.transform(item)
        return item

##Load trained GoogLeNet and ResNet50###
print('Load trained GoogLeNet and ResNet50 for segmentation prediction')
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
#Load trained Resnet50
PATH = '../TrainedModel/Resnet50EP100.pth'
Res50model = models.resnet50(pretrained=True) #load resnet50.
num_ftrs = Res50model.fc.in_features

# Here the size of each output sample is set to 2.
# Alternatively, it can be generalized to nn.Linear(num_ftrs, len(class_names)).
Res50model.fc = nn.Linear(num_ftrs, 5)

Res50model = Res50model.to(device)
Res50model.load_state_dict(torch.load(PATH))
#Res50model.eval()

#Load trained GoogleNet
PATH = '../TrainedModel/GoogLeNetEP100.pth'
GLNmodel = models.googlenet(pretrained=True) #load resnet50.
num_ftrs = GLNmodel.fc.in_features

# Here the size of each output sample is set to 2.
# Alternatively, it can be generalized to nn.Linear(num_ftrs, len(class_names)).
GLNmodel.fc = nn.Linear(num_ftrs, 5)

GLNmodel = GLNmodel.to(device)
GLNmodel.load_state_dict(torch.load(PATH))
#Res50model.eval()

def plotSegmentation(img, seg, saveName):
    True_names = ['hard_disk', 'disk', 'chip', 'reader', 'y_part']
    pred_names = ['chip', 'disk', 'hard_disk', 'reader', 'y_part']
    RGBImg = np.moveaxis(np.uint8(img[0:3]), 0, -1) #Reshape channeL from [C, H, W] to [H, W, C]
    plt.imshow(RGBImg)
    predArr = []
    for i in range(5):
        color = cv2.cvtColor(seg[i], cv2.COLOR_GRAY2BGRA)
        color[:,:,0] = color[:,:,0]*random.randint(1,255)
        color[:,:,1] = color[:,:,1]*random.randint(1,255)
        color[:,:,2] = color[:,:,2]*random.randint(1,255)
        color[color[:,:,0]==0, 3] = color[color[:,:,0]==0, 3]-255
        color4pred = cv2.cvtColor(seg[i], cv2.COLOR_BGRA2BGR)
        color4pred = np.moveaxis(color4pred, -1, 0) #Reshape channeL from [H, W, C] to [C, H, W]
        predArr.append(color4pred)
        plt.imshow(color, alpha=0.5)
    plt.axis('off')
    plt.show()
    predArr = torch.tensor(np.array(predArr)).to(device).float()
    
    #Resnet50 predict results
    print('Resnet50 prediction:')
    outputs = Res50model(predArr)
    m = nn.Softmax(dim=1)
    proOutput = m(outputs)
    _, preds = torch.max(outputs, 1)
    pro = []
    idx = [2, 1, 0, 3, 4]
    for j in range(outputs.size()[0]):
        pro.append(proOutput[j][idx[j]])

    for j in range(outputs.size()[0]):
        #print('True:%s Pred: %s with %s probabilities' %(True_names[j], class_names[preds[j]], round(pro[j].item(),4)))
        print('%s with prediction%s probabilities' %(True_names[j], round(pro[j].item(),4)))
        
    #GoogLeNet predict results
    print('GoogLeNet prediction:')
    outputs = GLNmodel(predArr)
    m = nn.Softmax(dim=1)
    proOutput = m(outputs)
    _, preds = torch.max(outputs, 1)
    pro = []
    idx = [2, 1, 0, 3, 4]
    for j in range(outputs.size()[0]):
        pro.append(proOutput[j][idx[j]])

    for j in range(outputs.size()[0]):
        #print('True:%s Pred: %s with %s probabilities' %(True_names[j], class_names[preds[j]], round(pro[j].item(),4)))
        print('%s with prediction%s probabilities' %(True_names[j], round(pro[j].item(),4)))
    #if not os.path.exists('checkSeg'):
        #os.makedirs('checkSeg')       
    #plt.savefig('./checkSeg/'+saveName)
```

```python
DatasetArr = np.load('../Dataset/DatasetArrV2.npy') #Rade the trianing data.

lengths = [round(len(DatasetArr)*0.9)-1, round(len(DatasetArr)*0.1)]
train, test = random_split(DatasetArr, lengths ,generator=torch.random.manual_seed(42)) #Shuffle data with random seed 42 before split train and test

trainData = [] 
for i in range(len(train)):
    trainData.append([torch.tensor([train[i][0]/255, train[i][1]/255, train[i][2]/255]), torch.tensor([train[i][3], train[i][4], train[i][5], train[i][6], train[i][7], train[i][8]])])
trainLoader = torch.utils.data.DataLoader(trainData, shuffle=True, batch_size=5)

testData = [] 
for i in range(len(test)):
    testData.append([torch.tensor([test[i][0]/255, test[i][1]/255, test[i][2]/255]), torch.tensor([test[i][3], test[i][4], test[i][5], test[i][6], test[i][7], train[i][8]])])
testLoader = torch.utils.data.DataLoader(testData, shuffle=False, batch_size=5)


```

```python
net = Sequential(
            # Defining 1st 2D convolution layer
            Conv2d(3, 64, kernel_size=3, stride=1, padding=1), #224@64
            BatchNorm2d(64),
            ReLU(inplace=True),
            MaxPool2d(kernel_size=2, stride=2),
            # Defining 2nd 2D convolution layer
            Conv2d(64, 128, kernel_size=3, stride=1, padding=1), #112@128
            BatchNorm2d(128),
            ReLU(inplace=True),
            MaxPool2d(kernel_size=2, stride=2),
            # Defining 3rd 2D convolution layer
            Conv2d(128, 256, kernel_size=3, stride=1, padding=1), #56@256
            BatchNorm2d(256),
            ReLU(inplace=True),
            MaxPool2d(kernel_size=2, stride=2),
            # Defining 4th 2D convolution layer
            Conv2d(256, 512, kernel_size=3, stride=1, padding=1), #28@512
            BatchNorm2d(512),
            ReLU(inplace=True),
            MaxPool2d(kernel_size=2, stride=2),
            # Defining 4th 2D convolution layer
            Conv2d(512, 256, kernel_size=3, stride=1, padding=1), #14@512
            BatchNorm2d(256),
            ReLU(inplace=True),
            MaxPool2d(kernel_size=2, stride=2),
            ConvTranspose2d(256, 256, 4, stride=2, padding=1), #14@512
            BatchNorm2d(256),
            ReLU(inplace=True),
            ConvTranspose2d(256, 128, 4, stride=2, padding=1), #28@256
            BatchNorm2d(128),
            ReLU(inplace=True),
            ConvTranspose2d(128, 64, 4, stride=2, padding=1), #56@64
            BatchNorm2d(64),
            ReLU(inplace=True),
            ConvTranspose2d(64, 32, 4, stride=2, padding=1), #112@32
            BatchNorm2d(32),
            ReLU(inplace=True),
            ConvTranspose2d(32, 6, 4, stride=2, padding=1), #224@16
            ReLU(inplace=True)
            #ReLU(inplace=True)
        )

net = net.to(device)
criterion = nn.MSELoss()
optimizer = optim.SGD(net.parameters(), lr=0.001, momentum=0.9)
```

```python
#%%time
trainTestLog = open("ColorizingSigmoidTrainTestLog1000.txt","w+") 
n_epoch = 10000
for epoch in range(n_epoch):  # loop over the dataset multiple times
#epoch = 0
#while True:
    train_loss = 0.0
    for i, data in enumerate(trainLoader, 0):
        inputs, labels = data[0].to(device), data[1].to(device) #Color
        inputs = inputs.float()
        labels = labels.float()
        # zero the parameter gradients
        optimizer.zero_grad()

        # forward + backward + optimize
        #print(inputs.shape)
        #print(labels.shape)
        outputs = net(inputs)
        #print(outputs.shape)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        # print statistics
        train_loss += loss.item()
    logStr = 'Train epoch: %d, MSE_Loss: %.10f' % (epoch + 1, train_loss/ (i+1))
    trainTestLog.write(logStr+'\n') #Save training records into log.
    print(logStr)
    epoch += 1

print('Finished Training')
```
